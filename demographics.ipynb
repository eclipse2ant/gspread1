```python
import pandas as pd

data = pd.read_csv('ma030000.csv',
                   encoding="shift-jis",
                   header=None)
data.head()
```

```python
data = pd.read_csv('ma030000.csv',
                   encoding="shift-jis",
                   skiprows=4,
                   header=None)
data.drop(data.tail(21).index, inplace=True)
data.head()
```

```python
# 先頭3行をcol_dataに読み込む
col_data = pd.read_csv('ma030000.csv',
                       encoding="shift-jis",
                       nrows=3,
                       header=None)

# 先頭1行を抜き出し、欠損値を行方向に前の値で埋める
col_head = col_data.head(1)
col_head = col_head.fillna(method="ffill", axis=1)

# col_dataの先頭1行目を書き換える
col_data.iloc[0] = col_head.iloc[0]

# colsに結合したカラム名を追加する
cols = []
for i in col_data.columns:
    col = ''.join(list(col_data[i].dropna()))
    cols.append(col)

# 0番目のカラム名を「都道府県」で置き換える
cols[0] = '都道府県'

# dataのカラム名をセットする
data.columns = cols
data.head()
```

```python
import numpy as np

# 欠損値「-」「・」「…」をnp.nanに置き換える
data.replace('-', np.nan, inplace=True)
data.replace('・', np.nan, inplace=True)
data.replace('…', np.nan, inplace=True)

# データ型をすべてfloat型に変換する
data.iloc[:, 2:] = data.iloc[:, 2:].astype(float)
data.dtypes
```

```python
import matplotlib.pyplot as plt

# 日本語が文字化けしないようにフォントを設定する
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Yu Gothic', 'Meiryo']

# グラフの大きさ（横の長さ, 縦の長さ）を決める
plt.rcParams['figure.figsize'] = (20, 10)

# X軸に都道府県名、Y軸に出生数を表示する
x = data['都道府県']
y = data['出生数']

# グラフのタイトル、各軸のラベルを設定する
plt.title('全国の出生数')
plt.xlabel('都道府県')
plt.ylabel('出生数')

# X軸の各項目を横書きから縦書きに変更する
plt.xticks(rotation=90)

# 棒グラフを描画する
plt.bar(x, y)
```

```python
data.to_csv('demographics.csv',
            index=False,
            encoding='utf_8_sig')
```
